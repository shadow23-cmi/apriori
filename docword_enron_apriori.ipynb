```python
## ***Warning ,may crash !!!!****##
## ***Trygoogle colab*** ##
import pandas as pd
import numpy as np
import time

url="https://github.com/shadow23-cmi/apriori/raw/master/docword_enron.txt"
```

```python
data =pd.read_csv(url,
                  skiprows=3,
                  header=None,
                  sep= " ",
                  names=["docId","wordId","count"])
```

```python
# not recomended for docword_enron.txt#
# it will crash the system #
# .pivot() or .unstack() does not work for big dataframe  #


data_clean=data.pivot(index = "docId",                               
                      columns = "wordId",
                      values = "count")


```

```python
#data_clean=data.groupby(["docId","wordId"])["count"].sum().unstack().reset_index().fillna(0).set_index("docId")
```

```python
data_clean=data_clean.fillna(0)
```

```python
data_clean=data_clean.astype(bool).astype(int)
```

```python

#set of all transactions
start_transaction_time=time.time()
Transactions=[]
for id in data_clean.index:
    ls=np.flatnonzero(data_clean.loc[id]).tolist()
    ls1=[item+1 for item in ls]
    ls2=set(ls1)
    Transactions.append(ls2)
end_transaction_time=time.time()
print(end_transaction_time-start_transaction_time)    

#candidate-gen function
def candidate_gen(f):
    c_set=[]
    c=set()
    for f1 in f:
        for f2 in f :
            if (((f2.difference(f1)).union(f1.difference(f2))=={max(f1),max(f2)})&(max(f1)<max(f2))):
                c =f1.union({max(f2)})
                c_set.append(c)
               
                for i in c:
                    if c.difference({i}) not in f:
                        
                        if (c_set != []) & (c in c_set):
                           
                            c_set.remove(c)
    
    return c_set
#converts the frequent itemset list to words vocubaliry file
def to_word_conversion(word_index):
    url1= "https://github.com/shadow23-cmi/apriori/raw/master/vocab_enron.txt"
    vocabulary_file=pd.read_csv(url1,header=None)
    vocabulary=vocabulary_file[0].values.tolist()
    frequent_set_words=[]
    for frequent_set in word_index:
        temp_word_set=set()
        for index in frequent_set:
            temp_word_set.add(vocabulary[index-1])
        frequent_set_words.append(temp_word_set)
    return frequent_set_words

#counts the frequency of a candidate set
def support_count_of_sets(input_set):
    count=0
    for t in Transactions:
        if input_set.issubset(t):
            count =count+1
    return count
    

# apriori function
# apriori function
def apriori(data_clean,min_sup,k):
    
    global F                           # defining global variabe as to be used in output_file function
    global K
    global total_time
    global final_word_list
    global final_f
    c1=[]
    f=[]
    no_of_transactions=max(data_clean.index)
    start_time=time.time()
    for item in data_clean.columns:
        c1.append(set([item]))
        
    for item in data_clean.columns:
        if ((data_clean[item].sum())/no_of_transactions)>=min_sup:
            f.append(set([item]))
    final_f=[]
    final_f=final_f+f
   
    
    for i in range(2,k+1):
        
        c_set=candidate_gen(f)
        
        temp=len(f)
        for item in c_set:
            
            if (support_count_of_sets(item) /no_of_transactions) >= min_sup:
                f.append(item)                           #  now f contains k-1,k length frequent item sets
                             
        f=f[temp:]                                       # this removes k-1 length frequent item sets from f
        final_f=final_f+f                                # appends all k length frequent items to f_final
    end_time  =time.time() 
    total_time=end_time-start_time
    F=min_sup
    K=k
    final_word_list=to_word_conversion(final_f)
    
    return final_f




def output_file_apriori(min_sup,k):
    apriori(data_clean,min_sup,k)
    output_file=open("output_kos.txt","a+")
    output=""
    output=output+"\n\n"+"F="+str(F)+"\n\n"+"K="+str(K)+"\n\n"+"Runtime is :"+str(total_time)+" seconds"+"\n\n"
    output=output+"The Frequent sets are:"+"\n\n"+"Word indices:"+"\n\n"
    output_file.write(output)
    output_file.write(str(final_f))
    output_file.write("\n\n"+"words:"+"\n\n")
    output_file.write(str(final_word_list))
    print("F=",min_sup)
    print("K=",k)
    print("Runtime is :"+str(total_time)+" seconds")
    print("The Frequent sets are:")
    print(final_word_list)
#############################################################################

#Give input as minimum support and k,k being the maxmimum length frequent itemset

output_file_apriori(0.8,10)

output_file_apriori(0.7,10)

output_file_apriori(0.6,10)

output_file_apriori(0.5,10)

output_file_apriori(0.4,10)

output_file_apriori(0.3,10)

output_file_apriori(0.2,10)

output_file_apriori(0.15,10)

output_file_apriori(0.1,10)
```
